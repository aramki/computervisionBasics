```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing OpenCV
import cv2
```

```python
#Reading the image; Coverts directly to ndarray
#Check the output for 'NoneType'
img = cv2.imread('./train.jpg')
```

```python
type(img)
```

```python
#Checking if the returned img is None; Break here if the image isn't found
if (img is None):
    print('Returned Null')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
#The output above looks weird as RGB of matplotlib isn't the same for openCV
#This should be fixed
#So, we need to perform a transformation
```

```python
#Transforming BGR (as expected by openCV) to RGB
fixedImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fixedImg)
```

```python
#Looking at grayscales
imgGray = cv2.imread('./train.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
#Given that we have converted it to grayscale, it is a 2x2 array now
imgGray.shape
```

```python
plt.imshow(imgGray)
#The darker values are bluish in the image below, while the lighter values are yellowish-green
```

```python
#To see only in grayscale (black/white)
plt.imshow(imgGray, cmap = 'gray')
```

```python
plt.imshow(fixedImg)
```

```python
fixedImg.shape
```

```python
#Resizing
#Note that the first number 360 is a conversion for 480 pixels, 
#while 240 is a conversion for 360 pixels
resizedImg = cv2.resize(fixedImg, (360, 240))
```

```python
plt.imshow(resizedImg)
```

```python
#Resizing by ratio
newratio_breadth = 0.5
newratio_height = 0.5

resizedByRatio = cv2.resize(fixedImg, (0,0), fixedImg, newratio_breadth, newratio_height)
```

```python
plt.imshow(resizedByRatio)
```

```python
resizedByRatio.shape
```

```python
fixedImg.shape
```

```python
#Flipping Images
#Horizontal Flip (along X axis)
flipImage = cv2.flip(fixedImg, 0) 
plt.imshow(flipImage)
```

```python
#Vertical Flip (along Y axis)
flipImage = cv2.flip(fixedImg, 1) 
plt.imshow(flipImage)
```

```python
#Both X & Y flip
flipImage = cv2.flip(fixedImg, -1) 
plt.imshow(flipImage)
```

```python
#Writing the image back to a file
cv2.imwrite('flip.jpg', flipImage)
```

```python
newImg = cv2.imread('./flip.jpg')
```

```python
plt.imshow(newImg)
```

```python
#Converting the BGR to RGB so that it can be viewed properly on other editors
fixedFlip = cv2.cvtColor(newImg, cv2.COLOR_RGB2BGR)
cv2.imwrite('flip.jpg', fixedFlip)
```

```python
#Using a canvas - figSize 10x8 inches
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(newImg)
```
